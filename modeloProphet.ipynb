```python
%pylab inline
import pandas as pd
import numpy as np
%matplotlib inline

from neuralprophet import NeuralProphet

from sklearn.metrics import mean_absolute_error
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_excel('/content/drive/My Drive/Sample - Superstore.xls',parse_dates=True)
```

```python
data.dtypes
```

```python
data.rename(columns={'Order Date': 'Order'}, inplace = True)
data.Sales.replace((',','.'),inplace =True)

data.head()
```

```python
pd.set_option("display.max_columns",21)
data.head()
```

```python
data['State'].value_counts().head(21)
```

```python
tmax = data[['Order','Sales']]
tmax = tmax.rename(columns= {'Order': 'ds', 'Sales': 'y'})

tmax['month'] = tmax['ds'].dt.to_period('M').dt.to_timestamp()
```

```python
tmax = tmax.groupby('month').agg('sum').reset_index()

tmax = tmax.rename(columns= {'month': 'ds'})
```

```python
tmax
```

```python
treino = tmax[:40]
teste = tmax[40:]
```

```python
treino
```

```python
m = NeuralProphet()
```

```python
metrics = m.fit(treino)
```

```python
pred = m.predict(teste)
```

```python
teste
```

```python
mean_absolute_error(pred['y'], pred['yhat1'])
```
